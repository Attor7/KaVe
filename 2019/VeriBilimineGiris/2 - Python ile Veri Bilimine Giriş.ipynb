```python
grades = pd.read_excel("learn.xls", encoding='utf-8')
```

```python
grades.columns = ['Last Name', 'First Name', 'Student ID', 'Final',
                 'Project ', 'midterm', 'Quiz2', 'Quiz1', 'Quiz3', 'Quiz3B', 
                 'Quiz3C', 'Quiz4', 'Quiz4B', 'Quiz5A', 'Quiz5B', 'Quiz5C']
```

## Eksik degerler

```python
grades.fillna(0, inplace=True)
grades.head()
```

```python
grades.info()
```

```python
grades.describe()
```

## Uygunsuz Degerleri degistirme

```python
grades = grades.replace({'In Progress':0})
grades = grades.replace({'Needs Grading':0})
```

```python
grades.describe()
```

## Yeni Oznitelik Ekleme

```python
all_quizes = ['Quiz2', 'Quiz1', 'Quiz3', 'Quiz3B', 'Quiz3C', 'Quiz4', 'Quiz4B', 'Quiz5A', 'Quiz5B', 'Quiz5C']
quizes = grades[all_quizes].sum(axis=1)

grades['quiz'] = quizes
grades['discussion'] = 5
```

## Oznitelik Silme

```python
grades.drop(all_quizes, axis=1, inplace=True)
```

```python
grades.head()
```

## Notlari ekle

```python
grades['grade'] = grades['Final'] * 0.4 + grades['midterm'] * 0.3 \
                + grades['Project '] * 0.15 + grades['quiz']  + grades['discussion'] 
```

```python
grades.head()
```

### Kalanlar
 - 35'in alti kaliyor

```python
sns.distplot(grades.grade, bins=100)
plt.axvline(35, 0, 1, color = "red")
```

```python
kalan = np.sum(grades.grade < 35)
N = len(grades)
kalan_oran = kalan / N
```

```python
kalan, kalan_oran, N, grades.grade.max()
```

 - Ogrencilerin %28i kaliyor

## Harf notlarini hesaplama

```python
letter_grade_cuts = [0, 35, 39, 43, 47, 51, 55, 62, 68, 80, 90, 100]
letters = ["F","D", "D+","C-", "C", "C+", "B-", "B", "B+", "A-", "A"]

grades['letters'] = pd.cut(grades.grade, letter_grade_cuts, right=False, labels =letters)
```

```python
grades.head()
```

## Sinifta kalmamak icin ne gerekiyor?

```python
grades['Fail'] = grades.letters == 'F'
```

```python
grades.head(2)
```

```python
sns.pairplot(x_vars=["midterm", "Project ", "quiz"], y_vars=["Final"], 
             hue="Fail", 
             data=grades, height=5, aspect=.8, kind="reg");
```

## En iyi ogrenciler arasinda olmak icin ne gerekiyor?

```python
grades['Best'] = grades.letters.apply(lambda x: x in ["B", "B+", "A-", "A"])
```

```python
sns.pairplot(x_vars=["midterm", "Project ", "quiz"], y_vars=["Final"], 
             hue="Best", 
             data=grades, height=5, aspect=.8, kind="reg");
```
