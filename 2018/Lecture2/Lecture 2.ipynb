## Gradyan Inise hazirlik

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def Hata(w):
    return ((w-3)**2)+5

def turev(w):
    return 2*(w-3)

def gradyan_inis(w = 0, alpha = 0.05, dongu = 20):
    W = np.zeros(dongu)
    
    for i in range(dongu):
        W[i] = w
        w = w - alpha * turev(w)

    return W
```

```python
a = 0.95
W = gradyan_inis(w = 40, alpha = 0.95)
W
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
t = np.arange(-50,50,0.5)

plt.plot(W,Hata(W),'k')
plt.scatter(W,Hata(W), color = 'red')
plt.plot(t,Hata(t))
plt.xlabel('w'); plt.ylabel('Hata(w)'); plt.title("alpha = " + str(a))
plt.grid()
```

```python
a = 0.05
W = gradyan_inis(w = 40, alpha = a)
t = np.arange(-50,50,0.5)

plt.plot(W,Hata(W),'k')
plt.scatter(W,Hata(W), color = 'red')
plt.plot(t,Hata(t))
plt.xlabel('w'); plt.ylabel('Hata(w)'); plt.title("alpha = " + str(a))
plt.grid()
```

## All in one place

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.linear_model import LinearRegression
```

```python
# Collect data
x = np.arange(10, 100, step= 5)
y = 2 * x + 5 + 20 * np.random.randn(18)

x = x.reshape(-1, 1)
```

```python
# train test split
x_test = np.arange(100)
x_test = x_test.reshape(-1, 1)
```

```python
# Apply Machine learning algorithm
model = LinearRegression()
# learn parameters with fit
model.fit(x,y)
# prediction
y_pred = model.predict(x_test)
```

```python
plt.plot(x_test, y_pred, 'r')
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("x")
```

```python
#Evaluation
# print the R-squared value for the model
model.score(x, y)
```

## Logistic Regression

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline 
from sklearn.linear_model import LogisticRegression
```

```python
# Collect data
x0 = np.random.randn(100) + 2
x1 = np.random.randn(100) + 2

x0_ = np.random.randn(100) + 2
x1_ = np.random.randn(100) + 2

xx0 = np.concatenate((x0,x0_))
xx1 = np.concatenate((x1,x1_))
y = np.concatenate((np.zeros(100) ,np.ones(100)))

d = {'x0': xx0, "x1":xx1, "y":y}
data = pd.DataFrame(d)

data.head(3)
```

```python
#data
```

```python
c=['b','r']
mycolors = [c[0] if i==0 else c[1] for i in y]
```

```python
data.plot.scatter('x0', 'x1' ,c=mycolors)
```

```python
# train test split
from sklearn.cross_validation import train_test_split
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Split dataset into train ab=nd test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 1/4, random_state = 0)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
y_test = y_test.reshape(-1,1)
```

```python
len(X_train), len(y_test)
```

```python
# Apply Machine learning algorithm
model = LogisticRegression()
# learn parameters with fit
model.fit(X_train, y_train)
# prediction
y_pred = model.predict(X_test)
```

```python
#Evaluation
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
