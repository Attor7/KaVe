# Soru2 Cevap 

```python
import pandas as pd
import numpy as np


isim_sutunu = []
soyisim_sutunu = []
numara_sutunu = []
ilknot_sutunu = []
ikincinot_sutunu = []
ortalama_sutunu = []

x = int(input("Kaç öğrenci gireceksin:"))
i = 0
while i<x:
    
    öğrenci_a = str(input("İsmi:"))
    isim_sutunu.append(öğrenci_a)
    öğrenci_b = input("Soyismi:")
    soyisim_sutunu.append(öğrenci_b)
    öğrenci_c = input("Numarası:")
    numara_sutunu.append(öğrenci_c)
    öğrenci_d = int(input("İlk notu:"))
    ilknot_sutunu.append(öğrenci_d)
    öğrenci_e = int(input("İkinci notu:"))
    ikincinot_sutunu.append(öğrenci_e)
    öğrenci_f = (öğrenci_d +öğrenci_e)/2
    ortalama_sutunu.append(öğrenci_f)
    i+=1


tablo = pd.DataFrame({"isim" : isim_sutunu , "soyisim" : soyisim_sutunu , "numara" : numara_sutunu , "ilknotu" : ilknot_sutunu , "İkincinotu" : ikincinot_sutunu , "ortalama" : ortalama_sutunu})
tablo.head()
```
